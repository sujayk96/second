```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = [x*0.1 for x in range(1000)]
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
y = [np.random.randint(100), np.random.randint(100), np.random.randint(100)]
x = ['a','b','c']
plt.bar(x,y)
plt.show()
```

```python
np.random.seed(55)
x = np.random.rand(1000)
plt.hist(x, bins = 10)
plt.show()
```

```python
x = [np.random.rand(10), np.random.rand(10)*5, np.random.rand(10)*10]
plt.boxplot(x)
plt.show()
```

```python
x = np.random.randn(10) + 35
y = x**2 - 10*x -50
plt.scatter(x,y)
plt.show()
```
