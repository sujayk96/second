```python
import pandas as pd
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pd.read_csv(url, names = names)
```

```python
data.corr()
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
scatter_matrix(data)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
datan = data.values
X = datan[:,0:8]
Y = datan[:,8]
```

```python
import numpy as np
scalar = StandardScaler().fit(X)
datar = scalar.transform(X)
np.set_printoptions(precision = 3)
print(datar[:10,:])
```

```python
data.head()
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

kFold = KFold(n_splits = 10, random_state = 7)
model = LogisticRegression(solver = 'liblinear')
results = cross_val_score(model, X, Y, cv = kFold)
print("Accuracy: {0} ({1})" .format(results.mean()*100.0, results.std()*100.0))

scoring = 'neg_log_loss'
results = cross_val_score(model, X, Y, cv = kFold, scoring = scoring)
print("Accuracy: {0} ({1})" .format(results.mean()*100.0, results.std()*100.0))

```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn import tree
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
#from sklearn.tree.export import export_text

scoring = 'accuracy'
kfold = KFold(n_splits=10, random_state=7)
#KNN
model = KNeighborsClassifier(n_neighbors = 3)
cv_results = cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
print("Accuracy: {0} ({1})" .format(cv_results.mean(), cv_results.std()))

model = svm.SVC(gamma = 'scale')
cv_results = cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
print("Accuracy: {0} ({1})" .format(cv_results.mean()*100.0, cv_results.std()*100.0))

model = tree.DecisionTreeClassifier()
cv_results = cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
print("Accuracy: {0} ({1})" .format(cv_results.mean()*100.0, cv_results.std()*100.0))
# = export_text(decision_tree, feature_names=data['feature_names'])

model = RandomForestClassifier(n_estimators=100, max_depth=3)
cv_results = cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
print("Accuracy: {0} ({1})" .format(cv_results.mean()*100.0, cv_results.std()*100.0))

model = SGDClassifier(max_iter=1000, tol=1e-3)
cv_results = cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
print("Accuracy: {0} ({1})" .format(cv_results.mean()*100.0, cv_results.std()*100.0))
```

```python
from sklearn.model_selection import GridSearchCV
import numpy as nd
```

```python
c_val = nd.array([10,1,0.1,0.01,0.001,0.0001])
parameters = {'C':c_val}
model = LogisticRegression(solver = 'liblinear')
#model = GridSearchCV(lr, parameters, cv=5)
grid = GridSearchCV(estimator=model, param_grid=parameters, cv=3)
grid.fit(X, Y)
print(grid.best_score_)
print(grid.best_estimator_.C)
```

```python
from sklearn.linear_model import Ridge
alphas = nd.array([1,0.1,0.01,0.001,0.0001,0])
param_grid = dict(alpha=alphas)
model = Ridge()
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid.fit(X, Y)
print(grid.best_score_)
print(grid.best_estimator_.alpha)
```

```python
import pickle
from sklearn.model_selection import train_test_split

url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pd.read_csv(url, names=names)
array = dataframe.values
X = array[:,0:8]
Y = array[:,8]
test_size= 0.33
X_train,X_test, Y_train,Y_test = train_test_split(X,Y,test_size=0.33,random_state=7)
model = LogisticRegression(solver='liblinear')
model.fit(X_train, Y_train)
# save the model to disk
filename = 'finalized_model.sav'
pickle.dump(model, open(filename, 'wb'))

# some time later...

# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, Y_test)
print(result)
```
